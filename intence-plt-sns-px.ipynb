```python
# импортируем библиотеки 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
# Загружаем произвольный Dataset
df = pd.read_csv('../input/daily-coffee-price/coffee.csv', index_col=0)
df.head()
```

## Seaborn

```python
sns.pairplot(data=df[:50], hue='Open');
```

```python
# Цена на закрытии торгов Close стабильно растёт в линейной зависимости к цене на открытиии Open
plt.figure(figsize=(15,10))
sns.scatterplot(data=df, x='Open', y='Close');
```

```python
# Посмотрим зависимости объёма закрытых слелок на момент закрытия торгов в срезе по датам
plt.figure(figsize=(15,10))
sns.scatterplot(data=df[:50], x='Close', y='Volume', hue='Date');
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=df[:100], x='Open', y='Close');
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=df[:50], x='Open', y='Close');
```

```python
plt.figure(figsize=(15,10))
heatmap=sns.heatmap(df[:50].corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Matrix', fontdict={'fontsize':14}, pad=12);
```

```python
plt.figure(figsize=(15,10))
sns.histplot(df[:100], x='High', y='Close');
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=df[:50], x='High', y='Close');
```
