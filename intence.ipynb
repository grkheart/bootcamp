### Matplotlib/Seaborn

#### Нарисуйте график sin(x), подпишите оси, добавьте сетку

```python
import matplotlib.pyplot as plt
import seaborn as sns

x=np.arange(-3.0, 3.0, 0.1)
y=np.sin(x)

fig=plt.subplots()
plt.plot(x,y, 'b-')
plt.grid()
plt.title('График функции sin(x)')

plt.ylabel('y')
plt.xlabel('x');
```

#### Нарисуйте график sin(x) и cos(x) на одном subplot'е и на разных, задайте цвета, которые вам больше нравятся

```python
x1=np.arange(-3.0, 3.0, 0.1)
y1=np.sin(x)
x2=np.arange(-3.0, 3.0, 0.1)
y2=np.cos(x)

fig=plt.subplots()
plt.plot(x1,y1, 'b-')
plt.plot(x2,y2, 'g-')

plt.grid()
plt.title('График функции sin(x)/cos(x)')

plt.ylabel('y')
plt.xlabel('x');
```

```python
x1=np.arange(-3.0, 3.0, 0.1)
y1=np.sin(x)
fig=plt.subplots()
plt.plot(x1,y1, 'b-')
plt.grid()
plt.title('График функции sin(x)')

x2=np.arange(-3.0, 3.0, 0.1)
y2=np.cos(x)
fig=plt.subplots()
plt.plot(x2,y2, 'y-')
plt.grid()
plt.title('График функции cos(x)')

plt.ylabel('y')
plt.xlabel('x')
plt.show();
```
